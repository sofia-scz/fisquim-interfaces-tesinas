```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('classic')
Title = 'RUN1-NN1'
```

```python
conv = pd.read_csv("convergence.dat",delim_whitespace=True, skiprows= 1, header = None )

plt.title(Title)
plt.xlabel('Epoch')
plt.ylabel(' MAE/ <RMSE>')
plt.yscale('log')
plt.grid(which='major',axis = 'y' , linestyle='--', color = 'k')
plt.plot(conv[0],conv[1], 'r', label = ' MAE ')
plt.plot(conv[0],conv[2], 'g', label = ' RMSE ')
plt.plot(conv[0],conv[3], 'r--', label = ' MAE Test ') 
plt.plot(conv[0],conv[4], 'g--', label = ' RMSE Test ') 
plt.legend()
plt.savefig('convergence-'+Title+'.jpg')
```

```python
enertest = pd.read_csv("energies.test",delim_whitespace=True,  header = None, usecols=[0,1])
enertrain = pd.read_csv("energies.train",delim_whitespace=True,header = None, usecols=[0,1])
Edft = -1.557925E+02
Eann = -1.557912E+02
enertrain[0] = enertrain[0]- Edft
enertrain[1] = enertrain[1]- Eann
enertest[0] = enertest[0]- Edft
enertest[1] = enertest[1]- Eann
plt.title(Title)
plt.xlabel('Energia DFT')
plt.ylabel('Energia ANN')
plt.axis('equal')
plt.scatter(enertrain[0],enertrain[1], color ='y', label = 'Trainset')
plt.scatter(enertest[0],enertest[1],color = 'g', label = 'Testset')
plt.plot(enertrain[0],enertrain[0], color = 'r')
plt.plot(enertrain[0],enertrain[0]+0.1, 'r--')
plt.plot(enertrain[0],enertrain[0]-0.1, 'r--')
plt.xlim(min(enertrain[0])-0.1, 5)
plt.ylim(min(enertrain[0])-0.1, 5)
plt.legend(loc = 'lower right')
plt.savefig('energyCorrelation-'+Title+'.jpg')
```
